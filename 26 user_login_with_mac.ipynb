```python
import pandas as pd
```

```python
first_half = pd.read_csv('../more_user_use_one_mac_1.csv', header=None, names=['分公司名称', '数据日期', '用户号码','大区id','分公司id','日期','mac','发送字节数','接收字节数','权限','月份'])
```

```python
second_half = pd.read_csv('../more_user_use_one_mac_2.csv', header=None, names=['分公司名称', '数据日期', '用户号码','大区id','分公司id','日期','mac','发送字节数','接收字节数','权限','月份'])
```

```python
df = pd.concat([first_half,second_half])
```

```python
df.columns
```

```python
more_accounts = df.groupby('mac').filter(lambda x: x['用户号码'].nunique() > 2)
```

```python
more_accounts.to_excel('../26 同一mac下有两个以上账号登录的清单.xlsx')
```

```python
more_accounts_num = more_accounts.groupby('mac')['用户号码'].nunique().sort_values(ascending=False)
```

```python
more_accounts_num.to_excel('../26 同一mac下有两个以上账号登录的数目.xlsx')
```
